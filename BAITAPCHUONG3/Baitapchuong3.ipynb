```python
#Bai10
print("=======================================\n")
print("CHUONG TRINH TINH BIEU THUC S(x,n)")
print("=======================================\n")
x = int(input("Nhap x: "))
n = int(input("Nhap n: "))
S = 0
for i in range(1, n + 1):
    S += x**i
    mau = 1
    for j in range(1, i + 1):
        mau *= j
    S += x**i / mau
print(f"S({x},{n}) = {S}")
print("=======================================\n")
print("CHUONG TRINH KET THUC!\nCHUC BAN MOT NGAY VUI!!!")
print("=======================================\n")
```

```python
#Bai18.1
print("=======================================\n")
print("CHUONG TRINH VE HINH VUONG RONG")
print("=======================================\n")
n = int(input("Nhap kich thuoc canh hinh vuong: "))
for i in range(n):  # lặp qua từng hàng
    if i == 0 or i == n - 1:
        # Hàng đầu tiên và cuối cùng: in toàn '*'
        print("* " * n)
    else:
        # Hàng giữa: '*' + khoảng trắng + '*'
        print("*" + "  " * (n - 2) + " *")
print("=======================================\n")
print("CHUONG TRINH KET THUC!\nCHUC BAN MOT NGAY VUI!!!")
print("=======================================\n")
```

```python
#Bai18.2
print("=======================================\n")
print("CHUONG TRINH VE HINH TAM GIAC VUONG")
print("=======================================\n")
n = int(input("Nhap kich thuoc canh hinh tam giac vuong: "))
for i in range(1, n + 1):  # i tăng từ 1 đến n
    # In khoảng trắng đẩy tam giác sang phải
    print("  " * (n - i), end="")  
    
    # In dấu '*'
    print("* " * i)
print("=======================================\n")
print("CHUONG TRINH KET THUC!\nCHUC BAN MOT NGAY VUI!!!")
print("=======================================\n")
```

```python
#Bai18.3
print("=======================================\n")
print("CHUONG TRINH VE HINH")
print("=======================================\n")
n = int(input("Nhap kich thuoc canh tam giac: "))

# ===== PHẦN 1: Hai tam giác rỗng ghép cạnh trái =====
for i in range(1, n + 1):
    if i < n:
        for j in range(i):
            if j == 0 or j == i - 1:  # In cạnh trái hoặc cạnh chéo
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
    else:
        # Dòng cuối in full sao
        for j in range(2 * n - 1):
            print("*", end=" ")
        print()

# ===== PHẦN 2: Tam giác ngược bên phải =====
for i in range(1, n):  # Có n-1 dòng
    spaces = 2 * n + 2 * (i - 1)  # Dịch sang phải chuẩn xác
    print(" " * spaces, end="")

    cols = n - i  # Số cột hiện tại
    for j in range(cols):
        if j == 0 or j == cols - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

print("=======================================\n")
print("CHUONG TRINH KET THUC!\nCHUC BAN MOT NGAY VUI!!!")
print("=======================================\n")
```

```python
#Bai19
import math
print("=======================================\n")
print("CHUONG TRINH TINH BIEU THUC S(x,n)")
print("=======================================\n")
x = float(input("Nhap x = "))
n = int(input("Nhap n = "))

# Biến lưu tổng
S = 0

# Vòng lặp duyệt số lẻ từ 1 -> 2n + 1
for i in range(1, 2 * n + 2, 2):
    S += x**i / math.factorial(i)  # x^i / i!

# In kết quả
print(f"S({x}, {n}) = {S}")
print("=======================================\n")
print("CHUONG TRINH KET THUC!\nCHUC BAN MOT NGAY VUI!!!")
print("=======================================\n")
```
